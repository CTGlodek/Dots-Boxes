```python
# Corey Glodek
#
# Dots and boxes
# 
# Main Program Agent

# import the support methods and classes
from board import *
from minimax import * 
from players import *

# bool for the main while loop
quit = False

while not quit:
    print("\n*************************************\nWelcome to Dots and Boxes Program.\n*************************************\n")

    #initializing the main score vars
    agent_score = 0
    my_score = 0

    #start the game
    game_board, ply = start_game()

    # build the board based on user input
    game_board.build_board()

    # show the initial state of the game board
    game_board.draw_board()

    # bool for the current instance of the game
    game_active = True

    while game_active:

        # Human playes move
        my_score = player(game_board, my_score, agent_score,ply)

        #check to see that moves are available, else end this instance
        game_active = game_board.moves_left()

        if not game_active:
            break

        #call the AI to make a move
        agent_score = agent(game_board,agent_score, my_score, ply)

        #check to see that moves are available, else end this instance
        game_active = game_board.moves_left()
        
        if not game_active:
            break

    #Exit this instance and show the final state
    print("The game has completed\n")
    game_board.draw_board()

    #state the sinner of the game
    winner(my_score, agent_score)

    #ask user if they would like to play again
    end_prog = input("\nWould you like to play again? \nPress Q to quit, any key to continue.")

    if end_prog =="Q":
         quit = True
```
